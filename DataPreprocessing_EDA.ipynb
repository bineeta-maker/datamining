**Part 1 - Solution #1.**

Given data-set : [3, 16, 20, 4, 2, 5, 10, 9, 13, 7, 14, 8].

Binning is grouping of a large data-set in a manner that the values are smoothed by their neighbouring values. As such the data-set needs to be sorted for the process.

Sorted data-set : [2, 3, 4, 5, 7, 8, 9, 10, 13, 14, 16, 20]

The following binning techniques are applied on the data, with 3 bins in each case.
1. **Equal-frequency binning:** The data is divided into 3 bin of same size(with given data set the size of the bins would be 4). As such the Bins are as follows:
*   Bin 1:[2, 3, 4, 5]
*   Bin 2:[7, 8, 9, 10]
*   Bin 3:[13, 14, 16, 20]

2. **Smoothing by bin boundaries:** After sorting, the data is divided in equal size bins and data in bin is replaced with the closest boundary values in the respective bin. As such the Bins are as follows:
*   Bin 1:[2, 2, 5, 5]
*   Bin 2:[7, 7, 10, 10]
*   Bin 3:[13, 13, 13, 20]

For Bin 1: \\
2, 5 are the boundary values. 3, 4 are replaced by 2, 5 respectively. 3 is closer to 2 and 4 is closer to 5. So the new bin values are calculated as 2, 2, 5, 5.

For Bin 2: \\
7, 10 are the boundary values. 8, 9 are replaced by 7,10 respectively. 8 is closer to 7 and 9 is closer to 10. So the new bin values are calculated as 7, 7, 10, 10.

For Bin 3: \\
13, 20 are the boundary values. 14, 16 are replaced by 13. 14, 16 are closer to 13 than 20. So the new bin values are calculated as 13, 13, 13, 20.

**Part 1 - Solution #2.**
Given data-set **A:[10, 5, 25, 50, 35]**
Below methods are used to normalize the following data.
1. **min-max normalization with min=0 and max=1**. The formula for min-max normalization is as below: \\
$v'_i = \frac{v_i-min}{max - min}\times(newmax- newmin) + newmin $ \\
where, $v_i = \text{values in A}$ \\
The new min-max values for the given data-set are:  \\
$v'_{10} = \frac{10-5}{50 - 5}\times(1- 0) + 0  = 0.3333$ \\
$v'_{5} = \frac{5-5}{50 - 5}\times(1- 0) + 0 = 0$ \\
$v'_{25} = \frac{25-5}{50 - 5}\times(1- 0) + 0 = 0.4444$ \\
$v'_{50} = \frac{50-5}{50 - 5}\times(1- 0) + 0 = 1$ \\
$v'_{35} = \frac{35-5}{50 - 5}\times(1- 0) + 0  = 0.6666$ \\
**Normalized data-set is [0.3333, 0, 0.4444, 1, 0.6666]**

2. **z-score normalization.** The formula for z-score normalization is as below: \\
$v'_i = \frac{v_i-\bar{A}}{\sigma A}$ \\
where, \\
$\bar{A}: \text{mean of A}$  \\
$\sigma A: \text{standard deviation of A}$ \\
Calculating the mean of data-set with the formula, the mean is: \\
$\bar{A} = \frac{sum\ of\ dataset}{Total\ number\ of\ dataset} =\frac{10+5+25+50+35}{5} = 25$  \\
Calculating the variance of data-set with the formula the standard deviation is: \\
$\sigma A =\sqrt{\frac{1}{N} \times \sum_{n=1}^{5}(x-\bar{x})^2} = \sqrt{\frac{(10-25)^2+(5-25)^2+(25-25)^2+(50-25)^2+(35-25)^2}{5}}$ \\
= $\sqrt{\frac{ 225+ 400 +0 +625+ 100}{5}} = \sqrt{270} = 16.43$ \\
The new z-Score values for the given data-set are:  \\
$v'_{10} = \frac{10-5}{16.43}  = -0.912$ \\
$v'_{5} = \frac{5-5}{16.43} = 1.217$ \\
$v'_{25} = \frac{25-5}{16.43} = 0$ \\
$v'_{50} = \frac{50-5}{16.43} = 1.521$ \\
$v'_{35} = \frac{35-5}{16.43}  = 0.608$ \\
**Normalized data-set is [-0.912, 1.217, 0, 1.521, 0.608]**

**Part 1 - Solution #3.**
Recorded Data: \\

|Rating/University |University A |University B |
|--------------- |---------- |---------- |
|Satisfied |71 |129 |
|Dissatisfied |37 |73 |

**Hypothesis $H_0$:** student satisfaction in  not correlated with a specific university and the feature are independent?

To evaluate the correlation between the event chi-square test would be used. The formula is: \\
$\chi^2 = \sum_{n=1}^{c} \sum_{n=1}^{r} \frac{(o_{ij} - e_{ij})^2}{e_{ij}}$ \\
where, \\
$o_{ij}:\text{observed frequency of the Event}$ \\
$e_{ij}:\text{expected frequency of the Event}$ \\
$c, r:\text{are the number of distinct value of the Event}$ \\

For the above case c=2, r=2 as there are two response(satisfied, unsatisfied) and two university(A,B).

The expected frequency is calculated as 
$e_{ij} = \frac{count a_i \times count b_i}{n}$ \\

where, \\
**a, b:** total number of tuple for the respective event. In the above case 'a' is being consider as response(satisfied, unsatisfied) and b are the two university(A,B) \\
**n:** total number of tuples for the recorded data. 

The Contingency Data: \\


|Rating/University |University A |University B | Total(Rating) |
|--------------- |---------- |---------- |---------- |
|Satisfied |71(69.6774) |129(130.3226) |200|
|Dissatisfied |37(38.3226) |73(71.6774) |110|
|Total(University)|108 |202 |310|

$expected_{satisfied,University A} = \frac{108 \times 200}{310} = 69.6774$ \\
$expected_{satisfied,University B} = \frac{202 \times 200}{310} = 130.3226$ \\
$expected_{dissatisfied,University A} = \frac{108 \times 110}{310} = 38.3226$ \\
$expected_{dissatisfied,University B} = \frac{202 \times 110}{310} = 69.6774$ \\

Chi-square for the give data-set is calculated as: \\
$\chi^2 = \frac{(71-69.6774)^2}{69.6774} + \frac{(129-130.3226)^2}{130.3226} + \frac{(37-38.3226)^2}{38.3226} +\frac{(73-71.6774)^2}{71.6774}$ \\
= 0.0251 +0.0456+0.0134+0.0244 \\
= 0.1085 \\


**Conclusion:** Since, the computed value 0.1085 is less than the given significance level 10.828, the hypothesis cannot be reject. It means that Student satisfaction is  not correlated with a specific university and they are independent.

```python
#import
import pandas as pd
import numpy as np

## assignment of Data from said csv file
data = pd.read_csv('country-income.csv', header='infer')
print('Actual Data : \n',data)

print('\n #1 Replace NaN values with the mean. \n')
## imputer to transform missing data NaN with mean of the column
from sklearn.impute import SimpleImputer
imp = SimpleImputer(missing_values=np.nan, strategy='mean')

## As the Imputer has been created for numeric NaN, Iterating through 
## the columns to identify the Float or Integer column datatype.
for column in data:
  if(data[column].dtype == np.float64 or data[column].dtype == np.int64):
    data[column] = (imp.fit_transform(data[[column]]).ravel()).round(1)
print('Transformed Data : \n',data)

print('\n #2 Replace categorical labels with numerical labels.')
from sklearn.preprocessing import LabelEncoder

print('\nColumn - Region and Online Shopper is used here as an example.')
print('Online Shopper: Yes/No are replaced with 1/0 resp.')
print('Region : Brazil, India, USA are replaced 0/1/2 resp.')
ord_enc = LabelEncoder()
data["Online Shopper"] = ord_enc.fit_transform(data["Online Shopper"])
data["Region"] = ord_enc.fit_transform(data["Region"])
print('\nTransformed Data : \n',data)
```

**Part 1 - Solution #6.**

```python
import pandas as pd
data = pd.read_csv('https://archive.ics.uci.edu/ml/machine-learning-databases/breast-cancer-wisconsin/breast-cancer-wisconsin.data', header=None)
data.columns = ['Sample code', 'Clump Thickness', 'Uniformity of Cell Size', 'Uniformity of Cell Shape',
                'Marginal Adhesion', 'Single Epithelial Cell Size', 'Bare Nuclei', 'Bland Chromatin',
                'Normal Nucleoli', 'Mitoses','Class']
data = data.replace('?',np.NaN)
data['Bare Nuclei'] = data['Bare Nuclei'].fillna(data['Bare Nuclei'].median())
data['Bare Nuclei'] = pd.to_numeric(data['Bare Nuclei'])
```

```python
import pandas as pd
from sklearn.decomposition import PCA
from sklearn import preprocessing
from scipy import stats

##  Removing the Class and Sample Code Column as they are identifiers and need to be normalized.
data_temp = data
data_temp = data_temp.drop('Class', axis=1)
data_temp = data_temp.drop('Sample code', axis=1)

## Standard variation and variance of each feature.
print('Standard deviation:\n', data_temp.std())
print('\nVariance:\n', data_temp.var())

## using z-Score, normalised the column data to able to analyse the data and visualise for analysis.
print('using z-Score, normalised the column data to able to analyse the data and visualise for analysis.') 
normalisedDF = pd.DataFrame()
for column in data_temp.columns:
  if (data_temp[column].dtype == np.int or data_temp[column].dtype == np.float):
    normalisedDF[column] = stats.zscore(data_temp[column])

## performing PCA and plotting the result as scattered plot.
numComponents = 2
pca = PCA(n_components=numComponents)
pca.fit(normalisedDF)

projected = pca.transform(normalisedDF)
projected = pd.DataFrame(projected,columns=['pc1','pc2'])
projected['Class'] = data['Class']
```

```python
import matplotlib.pyplot as plt

colors = {2:'b', 4:'r'}
markerTypes = {2:'+', 4:'x'}

for dClass in markerTypes:
    d = projected[projected['Class']==dClass]
    plt.scatter(d['pc1'],d['pc2'],c=colors[dClass],s=60,marker=markerTypes[dClass])

plt.ylim(-3, 20)
plt.show()
```


**Conclusion:**
1. With variance of 13.118368 for Bare Nuclei, it can be inferred that the data collected for this partular feature has a lot of difference and each data may not be identical or near to mean. The data collected is spread out.
2. The variance of Mitoses, Single Epithelial Cell Size, Bland Chromatin was found to be 2.941492, 4.903124, 5.945620 resp. The data collected for these features are expected to nearer to mean data.
3. For rest of the column the data is somewhat evenly spread across as compared to Bare Nuclei or Mitoses or Single Epithelial Cell Size or Bland Chromatin.
3. From the scatter plot it can be inferred that for Class 2 category, there few sample that are outlier for the sample data-set and with Bare Nuclei having the highest deviation of 13.118368, it contributes as feature for the outlier.

**Feature Scaling:** A feature in the dataset is defined by its magnitude and unit. The unit differs with the type of feature are chosen. E.g., to understand physical feature of human, features like height, weight, shoe size etc. are used. As such units used for them are meters, kilograms, natural number respectively. The magnitude quantifies the unit for the feature.

For multidimensional dataset where there are many features and sample for the feature, the measurement can vary, be large and can be huge for computation. In order to understand the feature, its distant or similar the concept of distance matrix is used to calculate. One of the commonly used is e.g. Euclidian distance, it is to calculate distance between the feature for many samples to find whether they are similar or distant. In such scenarios computing very large and different kind of measuremnts/weights of samples may take time. Feature scaling comes into picture where the magnitude is drill down to a standard value for the computation to take less time. 

For algorithm like Linear regression, KNN where distance are calculated with Euclidian, Minkowski , it is necessary for the data to be scaled to a simple format.
There are numerous ways for scaling:  
1. Min-Max Normalization: where the dataset of the feature is normalised to maximum of 1 and minimum of 0.
2. Z-Score Normalization: where the dataset of the feature is normalised by the mean and standard deviation.
3. Decimal scaling: where the dataset of the feature is normalized by moving the decimal point of values.

**Example:** For image processing the image is divided into pixels and each pixel is defined by it color or RGB value. The RGB value ranges from 0 to 255.

**Part 2 - Solution #5.**

Observation for heat map created through nearest neighbour interpolation. For the function $f: \mathbb{R}^2 \to \mathbb{R}$ given by $f(x,y) = z = x^2 + y^2$ , and the function would be computed with x, y : {10, 20, 30, ... 100}

**Observation found:** 

From the heatmap generated it can deduced that with increase in value of x or y the computed function value also increases for a given set of positively ordered natural numbers. If x, y are the least positive value then the f(x,y) will be smallest in the codomain z and f(x,y) is largest in the codomain z, if x, y are the highest positive values.

**For example:** 
 1. if y is taken as 10 and x is increased through the given range then the computed value increases accordingly.

$f(10,10) = 10^2 + 10^2 = 200$ \\
$f(20,10) = 20^2 + 10^2 = 500$ \\
$f(30,10) = 30^2 + 10^2 = 1000$ \\
$f(100,10) = 100^2 + 10^2 = 10100$ \\

2. The reverse is also applicable, if x is constant and y is increased through the range, then the computed value increases.

$f(10,10) = 10^2 + 10^2 = 200$ \\
$f(10,20) = 20^2 + 10^2 = 500$ \\
$f(10,30) = 100^2 + 10^2 = 10100$ \\
$f(10,100) = 10^2 + 100^2 = 10100$ \\

3. If x, y are the taken as 10, then z = 100 is the smallest value in the result data-set, if x, y are 100, then z = 10000 is the largest value in the result data-set.

$f(10,10) = 10^2 + 10^2 = 100$ is the smallest value in result data-set z \\
$f(100,100) = 10^2 + 10^2 = 10000$ is the largest value in result data-set z \\


```python
import numpy as np
x_range = np.linspace(10, 100, 10)
y_range = np.linspace(10, 100, 10)

# meshgrid: X[i, j] == x_range[j] and Y[i, j] == y_range[i]
X, Y = np.meshgrid(x_range, y_range)

# Z[i, j] == f(x_range[j], y_range[i])
Z = X**2 + Y**2

# Dataset representation
df = pd.DataFrame({'x': X.reshape(-1), 'y': Y.reshape(-1), 'z = f(x,y)': Z.reshape(-1)})
display(df.info())
3
# Interpolation: point (x, y) is colored according to the value z of the nearest point in the dataset
plt.imshow(Z, cmap='Blues', aspect='equal', interpolation='nearest')
plt.colorbar()

# xticks and yticks would show Z matrix indices
plt.xticks([])
plt.yticks([])

plt.show()
```
